```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.colors as mcolors
```

```python
# Read the CSV file
file_path = "/Users/angusmcmillan/Library/CloudStorage/Dropbox/UMAP/Angus_library_UMAP_results.csv"  # Replace with your CSV file path
data = pd.read_csv(file_path)
```

```python
# Define custom colors for each library value
custom_colors = {0: '#000000', 1: '#B6758A', 2: '#ACAEC2', 3: '#676C97', 4: '#5B927D'}

# Create a colormap based on the custom colors
cmap = mcolors.ListedColormap([custom_colors[val] for val in sorted(custom_colors.keys())])

for method in ["data"]:
    plt.figure(figsize=(8, 8))
    
    # Remove the frame
    sns.set_style("white")
    
    ax = sns.scatterplot(data=data,
                         x=f"UMAP_0",
                         y=f"UMAP_1",
                         hue="Library",
                         palette=custom_colors,  # Use the custom_colors palette
                         s=0.1,
                         alpha=0.5)
    
    # Set background transparent
    plt.gca().patch.set_alpha(0.0)
    
    # Move axis labels closer to the axis
    ax.xaxis.set_label_coords(0.5, -0.025)
    ax.yaxis.set_label_coords(-0.025, 0.5)
    
    # Remove the black frame
    for spine in ax.spines.values():
        spine.set_visible(False)

    # Save the figure with a transparent background
    plt.savefig(f"{method}_embedding_libraries.png", dpi=300, transparent=True)
    
    plt.show()
```
